```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_raw=pd.read_csv('population_total.csv')
df_raw
```

```python
df_pivot=df_raw.pivot_table(index='year', columns='country', values='population')
df_pivot
```

```python
df_pivot_selected=df_pivot[['Afghanistan','Argentina','Vietnam','China','Brazil','Canada','Denmark','France','Japan','Russia','Turkey','United Kingdom','United States']]
```

```python
df_pivot_selected
```

```python
df_pivot_selected.plot()
```

```python
# df_pivot_selected.pop('China')
# df_pivot_selected.plot()
```

```python
df_pct_change=df_pivot_selected.pct_change()*100
df_pct_change.plot(kind='line',xlabel='Year',ylabel='Population Growth Rate', title='Popualtion Growth Rate% by Country',figsize=(15,10))
```
